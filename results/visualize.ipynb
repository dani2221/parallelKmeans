```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
data_cen = []
for i in range(1,17):
    obj = {}
    folder = 'outputs-centroids/'
    if f'serial_{i}.json' in os.listdir(folder):
        f = open(f'{folder}serial_{i}.json')
        data = json.load(f)
        obj['serial'] = data[-1]['time']
    if f'centroid_{i}.json' in os.listdir(folder):
        f = open(f'{folder}centroid_{i}.json')
        data = json.load(f)
        obj['centroid'] = data[-1]['time']
    if f'batch_{i}.json' in os.listdir(folder):
        f = open(f'{folder}batch_{i}.json')
        data = json.load(f)
        obj['batch'] = data[-1]['time']
    if f'centroidbatch_{i}-2.json' in os.listdir(folder):
        f = open(f'{folder}centroidbatch_{i}-2.json')
        data = json.load(f)
        obj['cb-2'] = data[-1]['time']
    if f'centroidbatch_{i}-3.json' in os.listdir(folder):
        f = open(f'{folder}centroidbatch_{i}-3.json')
        data = json.load(f)
        obj['cb-3'] = data[-1]['time']
    if f'centroidbatch_{i}-4.json' in os.listdir(folder):
        f = open(f'{folder}centroidbatch_{i}-4.json')
        data = json.load(f)
        obj['cb-4'] = data[-1]['time']
    data_cen.append(obj)
data_cen
```

```python
data_sp = []
for i in [0.01, 0.1, 0.25, 0.5, 0.75, 1]:
    obj = {}
    folder = 'outputs-data/'
    if f'serial_5_{i}.json' in os.listdir(folder):
        f = open(f'{folder}serial_5_{i}.json')
        data = json.load(f)
        obj['serial'] = data[-1]['time']
    if f'centroid_5_{i}.json' in os.listdir(folder):
        f = open(f'{folder}centroid_5_{i}.json')
        data = json.load(f)
        obj['centroid'] = data[-1]['time']
    if f'batch_5_{i}.json' in os.listdir(folder):
        f = open(f'{folder}batch_5_{i}.json')
        data = json.load(f)
        obj['batch'] = data[-1]['time']
    if f'centroidbatch_5-2_{i}.json' in os.listdir(folder):
        f = open(f'{folder}centroidbatch_5-2_{i}.json')
        data = json.load(f)
        obj['cb-2'] = data[-1]['time']
    if f'centroidbatch_5-3_{i}.json' in os.listdir(folder):
        f = open(f'{folder}centroidbatch_5-3_{i}.json')
        data = json.load(f)
        obj['cb-3'] = data[-1]['time']
    data_sp.append(obj)
data_sp
```

```python
df_cen = pd.DataFrame(index=range(1,17), data=data_cen).interpolate(limit=1)
df_sp = pd.DataFrame(index=['1%', '10%', '25%', '50%', '75%', '100%'], data=data_sp).interpolate(limit=1)
```

```python
df_cen
```

```python
ax = df_cen.plot.line()
ax.set_xlabel('# of centroids')
ax.set_ylabel('time (s)')
plt.show()
```

```python
ax = df_sp.plot.line()
ax.set_xlabel('% of data')
ax.set_ylabel('time (s)')
plt.show()
```
